```python
from ultralytics import YOLO

modified_model = YOLO(
   '/homes/olagunju/ultralytics-local/ultralytics/cfg/models/11/custom-yolo-NOP5-CSM-ECA.yaml'
)

modified_model.train(
    data    = "/homes/olagunju/wheat_grading_project/Roboflow_data/Roboflow/550-images-dataset/data_config.yaml",
    epochs  = 400,
    imgsz   = 768,
    batch   = 16,
    seed    = 0,
    device  = "cuda:0",
    optimizer    = "SGD",
    lr0          = 0.01,
    momentum     = 0.937,
    weight_decay = 0.0005,
    conf    = 0.25,
    iou     = 0.45,
    max_det = 1500,
    cos_lr       = True,
    fliplr    = 0.5,
    flipud    = 0.0,
    degrees   = 10.0,
    translate = 0.1,
    scale     = 0.2,
    shear     = 2.0,
    hsv_h     = 0.01,
    hsv_s     = 0.1,
    hsv_v     = 0.1,
    mosaic    = 0.4,
    copy_paste = 0.0,
    project = "/homes/olagunju/wheat_grading_project/Roboflow_data/Roboflow/CLAUDE/Results",
    name    = "03-modified-NOP5-CSM-ECA-yolo11"
)
```

```python
# Load the best model from your results folder
best_model = YOLO('/homes/olagunju/wheat_grading_project/Roboflow_data/Roboflow/CLAUDE/Results/03-modified-NOP5-CSM-ECA-yolo11/weights/best.pt')

# Run validation specifically to see the top metrics again
metrics = best_model.val()
print(f"Best Recall: {metrics.results_dict['metrics/recall(B)']}")
```
